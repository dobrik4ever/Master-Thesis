```python
from skimage import io, measure, filters
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
fpath = '/Users/dobrik4ever/Library/CloudStorage/OneDrive-Personal/Documents/Study/winter_21-22/Master Thesis/4 Raw data/MAX_Marie2.tif'
img = io.imread(fpath)
print(img.shape)
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.title(fpath.split('/')[-1])
plt.show()
```

```python
gaus = filters.difference_of_gaussians(img,5, 7)
gaus[gaus < np.average(gaus)] = 0
gaus[gaus > np.average(gaus)] = 1
gaus = gaus *-1
plt.imshow(img, cmap='viridis')
plt.imshow(gaus, cmap='binary', alpha=.1)
plt.colorbar()
plt.axis('off')
plt.show()
```
